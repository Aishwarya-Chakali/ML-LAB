## Handling Missing Values

```python
#Replace the msiing values in Embarked column by 'S'
# Replace missing values in the 'Embarked' column with 'S'
train_data['Embarked'].fillna('S', inplace=True)

# Verify if the missing values are replaced
print("Number of missing values in 'Embarked' after replacement:", train_data['Embarked'].isnull().sum())

```

```python
#Impute the missing values in Fare column with median value
# Calculate the median value of the 'Fare' column
fare_median = train_data['Fare'].median()

# Replace missing values in the 'Fare' column with the median value
train_data['Fare'].fillna(fare_median, inplace=True)

# Verify if the missing values are replaced
print("Number of missing values in 'Fare' after replacement:", train_data['Fare'].isnull().sum())

```

```python
#Impute the Age column. It has too many missing values but it may be important. 
#We fill the missing values with numbers randomly drawn from distribution
#that has same mean and standard deviation as non missing Age data
import numpy as np

# Calculate the mean and standard deviation of the 'Age' column (excluding NaN values)
age_mean = train_data['Age'].mean()
age_std = train_data['Age'].std()

# Generate random values with the same mean and std for the missing entries
random_ages = np.random.normal(loc=age_mean, scale=age_std, size=train_data['Age'].isnull().sum())

# Assign these random values to the missing 'Age' entries
train_data.loc[train_data['Age'].isnull(), 'Age'] = random_ages

# Verify if the missing values are replaced
print("Number of missing values in 'Age' after replacement:", train_data['Age'].isnull().sum())
```

## Feature Engineering - Creating new features

```python
# Create and add a new feature FamilySize as a combination of SibSp and Parch
#FamilySize = Parch + SibSp + 1
# Create the FamilySize feature
train_data['FamilySize'] = train_data['SibSp'] + train_data['Parch'] + 1

# Display the first few rows to confirm the new feature
print(train_data[['SibSp', 'Parch', 'FamilySize']].head())
```

```python
# Create new feature IsAlone from FamilySize
#If FamilySize == 1 then IsAlone = True else IsAlone = False
# Create the IsAlone feature
train_data['IsAlone'] = train_data['FamilySize'] == 1

# Convert the boolean values to integers (optional, depending on your use case)
train_data['IsAlone'] = train_data['IsAlone'].astype(int)

# Display the first few rows to confirm the new feature
print(train_data[['FamilySize', 'IsAlone']].head())
```

## Model Building

```python
#Prepare X and y
# Prepare the feature matrix (X) and target vector (y)
X = train_data.drop('Survived', axis=1)  # Features (all columns except 'Survived')
y = train_data['Survived']              # Target variable

# Display the shapes of X and y to confirm
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

```python
# Label encoding of categorical features
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
label_encoder = LabelEncoder()

# List of categorical columns to encode
categorical_columns = ['Sex', 'Embarked']

# Apply label encoding to each categorical column
for column in categorical_columns:
    train_data[column] = label_encoder.fit_transform(train_data[column])

# Display the first few rows to verify the encoding
print(train_data.head())

```

```python
# concat df_categorical with original df



# Use pd.get_dummies to one-hot encode categorical features
df_categorical = pd.get_dummies(train_data[['Sex', 'Embarked']], drop_first=True)

# Concatenate the original data with the encoded columns
train_data = pd.concat([train_data.drop(['Sex', 'Embarked'], axis=1), df_categorical], axis=1)

# Display the updated DataFrame
print(train_data.head())

```

```python
#Do a pairplot of all features
import seaborn as sns
import matplotlib.pyplot as plt

# Create a pairplot of all features
sns.pairplot(train_data, hue='Survived', diag_kind='kde')

# Show the plot
plt.show()
```

```python
# Splitting the data into train and validation set - 90% train 10% validation
#X_train, X_test, y_train, y_validation
from sklearn.model_selection import train_test_split

# Splitting the dataset into train and validation sets
X_train, X_validation, y_train, y_validation = train_test_split(
    X, y, test_size=0.1, random_state=42, stratify=y
)

# Display the shapes of the resulting splits
print("Shape of X_train:", X_train.shape)
print("Shape of X_validation:", X_validation.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_validation:", y_validation.shape)
```
